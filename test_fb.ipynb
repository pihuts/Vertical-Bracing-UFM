```python
# filepath: api.py
from fastapi import FastAPI
from pydantic import BaseModel
from steel_lib.calculations import BoltShearCalculator  # Import your calculators
from steel_lib.data_models import BoltConfiguration, DesignLoads
from steel_lib.materials import BOLT_GRADES
import math
from steel_lib.si_units import si

app = FastAPI()

class CalculationRequest(BaseModel):
    bolt_diameter: float  # in inches
    bolt_grade: str  # e.g., "a325_x"
    n_shear_planes: int
    n_bolts: int
    # Add other fields as needed for your calculations

@app.post("/calculate/bolt_shear")
def calculate_bolt_shear(request: CalculationRequest):
    # Example: Set up configuration and calculator
    config = BoltConfiguration(
        bolt_diameter=request.bolt_diameter * si.inch,
        bolt_grade=BOLT_GRADES[request.bolt_grade],
        # Add other config fields from your models
    )
    calculator = BoltShearCalculator(connection=config)  # Adjust based on your calculator's init
    capacity = calculator.calculate_capacity(request.n_shear_planes, debug=False)
    return {"capacity": capacity}
```
